```python
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline
from huggingface_hub import login

#hugging face token

# Load the fine-tuned FinBERT model and tokenizer
finbert = BertForSequenceClassification.from_pretrained('ProsusAI/finbert', num_labels=3)
tokenizer = BertTokenizer.from_pretrained('ProsusAI/finbert')
```

```python
!pip install huggingface_hub
```

```python
nlp_pipeline = pipeline("sentiment-analysis",  model =  finbert, tokenizer = tokenizer)

# Example sentences related to Indian stock market news
sentences = [
    "The Indian stock market experienced a surge in trading activity.",
    "Investors are optimistic about the future of Indian financial markets.",
    "Concerns about economic uncertainties are affecting stock prices in India.",
    "Earnings reports from Indian companies show a positive trend."
    "The company recorded its highest-ever quarterly crude steel production at 7.03 million tonnes during the period. Saleable steel sales stood at 6.71 million tonnes, contributing to the company's operational performance."


]
```

```python
results = nlp_pipeline(sentences)
print(results)
```

```python
document = " ".join(sentences)

# Truncate the document if it's too long
max_length = 512  # FinBERT's maximum input length
tokens = tokenizer.tokenize(document)
if len(tokens) > max_length - 2:  # Account for [CLS] and [SEP] tokens
    tokens = tokens[:max_length - 2]
truncated_document = tokenizer.convert_tokens_to_string(tokens)
```

```python
# Perform sentiment analysis on the entire document
results = nlp_pipeline(truncated_document)
print(results)
```
