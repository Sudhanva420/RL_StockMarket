```python
import requests
import csv
from datetime import datetime

# Set your API key
API_KEY = "4DCU28MHOXYJZJ0X."

# Base URL for Alpha Vantage
BASE_URL = "https://www.alphavantage.co/query"

# Function to fetch news articles
def fetch_news(ticker, start_date, end_date, limit=1000):
    # Convert dates to Alpha Vantage's format: YYYYMMDDTHHMM
    start_date = start_date.strftime("%Y%m%dT%H%M")
    end_date = end_date.strftime("%Y%m%dT%H%M")

    # API parameters
    params = {
        "function": "NEWS_SENTIMENT",
        "tickers": ticker,
        "time_from": start_date,
        "time_to": end_date,
        "limit": limit,
        "apikey": API_KEY
    }

    # Make the API request
    response = requests.get(BASE_URL, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: Unable to fetch data (Status code: {response.status_code})")
        return None

# Parameters
ticker = "HDB"
start_date = datetime(2020, 1, 1)  # Start date
end_date = datetime(2022, 3, 7 )  # End date

# Fetch news articles
news_data = fetch_news(ticker, start_date, end_date)

# Save the result to a CSV file
if news_data and 'feed' in news_data:
    with open(f"{ticker}_news.csv", mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)

        # Write CSV header
        writer.writerow([
            "Title", "URL", "Time Published", "Authors", "Summary", "Banner Image",
            "Source", "Category", "Source Domain", "Overall Sentiment Score",
            "Overall Sentiment Label", "Ticker", "Ticker Relevance Score", "Ticker Sentiment Score", "Ticker Sentiment Label"
        ])

        # Write data rows
        for article in news_data['feed']:
            time_published = datetime.strptime(article['time_published'], "%Y%m%dT%H%M%S").strftime("%Y-%m-%d %H:%M:%S")
            authors = ", ".join(article.get('authors', []))
            ticker_data = article.get('ticker_sentiment', [{}])[0]  # Extract first ticker sentiment entry

            writer.writerow([
                article['title'],
                article['url'],
                time_published,
                authors,
                article.get('summary', 'N/A'),
                article.get('banner_image', 'N/A'),
                article.get('source', 'N/A'),
                article.get('category_within_source', 'N/A'),
                article.get('source_domain', 'N/A'),
                article.get('overall_sentiment_score', 'N/A'),
                article.get('overall_sentiment_label', 'N/A'),
                ticker_data.get('ticker', 'N/A'),
                ticker_data.get('relevance_score', 'N/A'),
                ticker_data.get('ticker_sentiment_score', 'N/A'),
                ticker_data.get('ticker_sentiment_label', 'N/A')
            ])

    print(f"CSV file '{ticker}_news.csv' has been created successfully.")
else:
    print("No news articles found or an error occurred.")
```

```python
import pandas as pd

# List of CSV file paths to concatenate
csv_files = [
    "/content/drive/MyDrive/files/HDB_news.csv",
    "/content/drive/MyDrive/files/HDB_news (1).csv",
    "/content/drive/MyDrive/files/HDB_news (2).csv",
    "/content/drive/MyDrive/files/HDB_news (3).csv",
    "/content/drive/MyDrive/files/HDB_news (4).csv",
    "/content/drive/MyDrive/files/HDB_news (5).csv"
]

# List to store DataFrames
df_list = []

# Read and append all CSV files
for file in csv_files:
    try:
        df = pd.read_csv(file)
        df_list.append(df)
        print(f"Successfully read {file}")
    except Exception as e:
        print(f"Error reading {file}: {e}")

# Concatenate all DataFrames into one
combined_df = pd.concat(df_list, ignore_index=True)

# Save the combined DataFrame to a new CSV file
output_file = "combined_output.csv"
combined_df.to_csv(output_file, index=False)

print(f"All specified CSV files have been combined and saved as '{output_file}'.")
```
